```python
# import an widget that allows a user to select a file
from ipyfilechooser import FileChooser
import pandas as pd
import FluorVis as fv

```

```python
# create a file chooser widget
fc = FileChooser()
# display the widget
display(fc)
# get the selected file
fc.selected
```

```python
# get the selected file path
fc.selected_path
# get the selected file name
fc.selected_filename
# create file name
fileName = fc.selected_path+"\\"+fc.selected_filename

# create dataframe
dfExcel = pd.read_excel(fileName, sheet_name=None, header=None) # gets all sheets

# show all sheet names
sheetNames = list(dfExcel.keys())
print("Sheet Names:")
print(sheetNames)
```

```python
import pygwalker as pyg

df = dfExcel['Quan. Result']
# drop columns 5-10
df = df.drop(df.columns[5:11], axis=1)

# make column header the names in the 0th row, and then drop the 0th row
df.columns = df.iloc[0]
df = df.drop(df.index[0])
# move instances of "Dye"="FAM" into a new df called dfFAM
dfFAM = df[df['Dye']=='FAM']
dfFAM_copy = dfFAM.copy()
# make a list of the replicate names from 1 to 8 in format of 'rep_1'
replicates = ['rep_'+str(i) for i in range(1,9)]
# add a column called "Replicates" that adds elements in "replicates". Loop rom the beginning to the end of dfFAM
dfFAM_copy['Replicates'] = [replicates[i%8] for i in range(len(dfFAM))]
# make a list of "dextranConcs" from 1 to 12 in floats
dextranConcs = [0.0, 0.6, 1.9, 2.5, 3.1, 3.8, 4.4, 5.0, 5.6, 6.3, 6.9,7.5]
# # add a column called "DextranConc" that adds elements in "dextranConcs". Loop rom the beginning to the end of dfFAM
dfFAM_copy['DextranConc'] = [dextranConcs[i%12] for i in range(len(dfFAM))]

display(dfFAM_copy)
```
